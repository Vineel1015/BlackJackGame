```python
# ESTABLISH GLOBAL VARIABLES

import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
# CREATE CARD CLASS

class Card():
    def __init__(self,rank,suit):
        self.rank = rank
        self.suit = suit
        self.value = int(values[rank])
    
    def __str__(self):
        return f'{self.rank} of {self.suit}'
    
    def ace_check(self):
        ace_value = int(input("Do you want your ace to be worth 1 or 11"))
        while True:
            if ace_value == 1:
                print("Your ace has been assigned a value of 1")
                return ace_value
            elif ace_value == 11:
                print("Your ace has been assigned a value of 11")
                return ace_value
            else:
                print("Please enter either a 1 or 11")
                continue

    def computer_ace_check(self):
        computer_choice = round(random.randint(0,1))
        if computer_choice == 0:
            return 1
        else:
            return 11
```

```python
# CREATE DECK CLASS

class Deck():
    
    def __init__(self):
        
        self.deck_of_cards = []
        
        for suit in suits:
            for rank in ranks:
                self.deck_of_cards.append(Card(rank, suit))

    def shuffle(self):
        return random.shuffle(self.deck_of_cards)
        
    def take_one_card(self):
        return self.deck_of_cards.pop(-1)
        
```

```python
# CREATE PLAYER CLASS

class Player():
    
    def __init__(self):
        self.hand = []    
        
    def show_hand(self):
        print("Player Hand:")
        for card in self.hand:
            print(f'Card {int(self.hand.index(card))+1} : {card.rank} of {card.suit}')
        
    def add_card(self, card):
        self.hand.append(card)
```

```python
# CREATE COMPUTER CLASS

class Computer():
    
    def __init__(self):
        self.hand = []
            
    def __str__(self):
        try:
            return f"{self.hand[0].rank} of {self.hand[0].suit} : This is the computer's first card\n"
        except:
            return f'No cards in hand!'     
```
